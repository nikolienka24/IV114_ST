# Method comparison plots - resources

### Load data

```python
csv_file = "statistics_all_datasets/combined_statistics.requirements.by_task.csv"  # replace with your CSV path
df = pd.read_csv(csv_file)
```

```python
sns.set(style="whitegrid")
```

### CPU time comparison

```python
plt.figure(figsize=(6, 4))
sns.barplot(x="Task", y="CPU_time_mean", data=df, hue="Task", palette="Blues_d", dodge=False, legend=False, errorbar=None)
plt.ylabel("CPU Time (mean, seconds)")
plt.title("CPU Time Comparison")
plt.tight_layout()
plt.savefig(f"figures/show/method_comparison/all_requirements/cpu_time_comparison.png",
            dpi=300, bbox_inches='tight')
plt.show()
```

### Wall Time Comparison

```python
plt.figure(figsize=(6, 4))
sns.barplot(
    x="Task", y="Wall_time_mean", data=df,
    hue="Task", palette="Greens_d", dodge=False, legend=False, errorbar=None
)
plt.ylabel("Wall Time (mean, seconds)")
plt.title("Wall Time Comparison")
plt.tight_layout()
plt.savefig(f"figures/show/method_comparison/all_requirements/wall_time_comparison.png",
            dpi=300, bbox_inches='tight')
plt.show()
```

### RAM Usage

```python
plt.figure(figsize=(6, 4))
sns.barplot(
    x="Task", y="RAM_used_MB_mean", data=df,
    hue="Task", palette="Oranges_d", dodge=False, legend=False, errorbar=None
)
plt.ylabel("RAM Used (MB, mean)")
plt.title("RAM Usage Comparison")
plt.tight_layout()
plt.savefig(f"figures/show/method_comparison/all_requirements/ram_usage_comparison.png",
            dpi=300, bbox_inches='tight')
plt.show()
```

### CPU Efficiency

```python
plt.figure(figsize=(6, 4))
sns.barplot(
    x="Task", y="CPU_efficiency_mean", data=df,
    hue="Task", palette="Purples_d", dodge=False, legend=False, errorbar=None
)
plt.ylabel("CPU Efficiency (mean)")
plt.title("CPU Efficiency Comparison")
plt.tight_layout()
plt.savefig(f"figures/show/method_comparison/all_requirements/cpu_efficiency_comparison.png",
            dpi=300, bbox_inches='tight')
plt.show()
```

### RAM Efficiency

```python
plt.figure(figsize=(6, 4))
sns.barplot(
    x="Task", y="RAM_efficiency_mean", data=df,
    hue="Task", palette="Greys_d", dodge=False, legend=False, errorbar=None
)
plt.ylabel("RAM Efficiency (mean)")
plt.title("RAM Efficiency Comparison")
plt.tight_layout()
plt.savefig(f"figures/show/method_comparison/all_requirements/ram_efficiency_comparison.png",
            dpi=300, bbox_inches='tight')
plt.show()
```
